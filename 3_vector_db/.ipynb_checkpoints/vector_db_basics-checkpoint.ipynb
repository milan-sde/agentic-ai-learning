```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.create_collection("new_articles")
```

```python
collection.add(
  ids=["id1","id2","id3","id4"],
  documents=[
    "Apple is leading in a smart phone game with iPhone sales up by 35%",
    "Tesla booked a minor porfit of 1 billion $ in Q2",
    "Apples are high in fiber, vitamin C, and various antioxidants",
    "Space X got NASA contract worth 10 billion $"
  ]
)
```

```python
#fetch the embeddings:
data = collection.get(
  include=["documents","embeddings","metadatas"]
)
```

```python

# print embeddings
for i, emb in enumerate(data["embeddings"]):
  print(f"\nDocument {i+1}")
  print(f"Text : ",data["documents"][i])
  print("Embeddings length: ",len(emb))
  print("First 10 values",emb[:10])
```

```python
results = collection.query(
  query_texts=["This is a query related to Elon Musk"],
  n_results=2
)
print(results)
```

```python
results = collection.query(
  query_texts=["I love iPad"],
  n_results=2
)
print(results)
```
