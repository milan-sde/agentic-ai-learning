```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
# llm = ChatGoogleGenerativeAI(model='gemma-4-31b-it')
llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash')
response=llm.invoke("Where is Nirma University??")
print(response.content)
```

```python
response=llm.invoke([
  ["system","You are a helpful assistant that answers in short one line"],
  ["human","Where is Nirma University??"]
])

print(response.content)
```

```python
prompt="Write a single line tagline for a gold ornaments making shop.Only output the tagline, nothing else."
llm=ChatGoogleGenerativeAI(
  model="gemini-2.5-flash",
  temperature=0
)
response = llm.invoke(prompt)
print(response.content)
```

```python
prompt="Write a single line tagline for a gold ornaments making shop.Only output the tagline, nothing else."
llm=ChatGoogleGenerativeAI(
  model="gemini-2.5-flash",
  temperature=1
)
response = llm.invoke(prompt)
print(response.content)
```

## grok model

```python
from langchain_groq import ChatGroq
```

```python
llm=ChatGroq(
  model="openai/gpt-oss-120b",
  reasoning_format="parsed"
)

messages=[
  (
    "system",
    "You are a helpful assistant that answers in short one line"
  ),
  (
    "human",
    "Where is Nirma University?"
  )
]

ai_msg = llm.invoke(messages)
print(ai_msg.content)
```
